```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
gapminder = pd.read_csv('boxdata_q.csv')
#gapminder.head(n=10)
```

```python
sns.catplot(x='Model', y='Avg_return', 
            data = gapminder,
            jitter = '0.25')
```

```python
sns.catplot(x = 'Model', 
            y = 'Avg_return', 
            data = gapminder,
            jitter = False,
            height = 4,
            aspect=1.5)
```

```python
sns.catplot(x='Model', y = 'Avg_return', 
            data = gapminder,
            kind = 'box',
            height = 4,
            aspect=1.5)
```

```python
# make boxplot with Catplot
sns.catplot(x='Model', 
            y='Avg_return',
            kind="box",
            data=gapminder,
            height=4,
            aspect=1.5)
# add data points to boxplot with stripplot
sns.stripplot(x='Model', 
              y='Avg_return',
              data=gapminder,
              alpha=0.3,
              jitter=0.2,
              color='k')
```

```python
sns.catplot(x='Model',
            y='Avg_return', 
            data=gapminder,
            height=4,
            aspect=1.5,
            kind='boxen')
```

```python
gapminder.boxplot(by="Model", column="Avg_return", grid=False)
```

```python
# make boxplot with Seaborn
bplot=sns.boxplot(y='Avg_return', x='Model', 
                 data=gapminder, 
                 width=0.3,
                 palette="colorblind")
 

# add stripplot to boxplot with Seaborn
bplot=sns.stripplot(y='Avg_return', x='Model', 
                   data=gapminder, 
                   jitter=True, 
                   marker='o', 
                   alpha=0.5,
                   color='black')
```

```python
# plot boxplot with seaborn
bplot=sns.boxplot(y='Avg_return', x='Model', 
                 data=gapminder, 
                 width=0.3,
                 palette="colorblind")
 
# add swarmplot
bplot=sns.stripplot(y='Avg_return', x='Model',
              data=gapminder, 
              #color='gray',
              palette="deep",
              alpha=.8)

bplot.axes.set_title("Performance on memory task",fontsize=16)
bplot.set_xlabel("LSTM variant",fontsize=14) 
bplot.set_ylabel("Avg return per episode",fontsize=14)
bplot.tick_params(labelsize=10)            
bplot.set_ylim([-4.5,8.5])  
```

```python
# save as jpeg
bplot.figure.savefig("boxplot_example_q.png", format='png', dpi=100, transparent=False, facecolor='w', edgecolor='w')
```
